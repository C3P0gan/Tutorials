# Data Science: análise de séries temporais

## 01. Tendência das vendas

```python
import pandas as pd
```

```python
alucar = pd.read_csv('data/alucar.csv')
alucar.head()
```

```python
print('Quantidade de linhas e colunas:', alucar.shape)
```

```python
print('Quantidade de dados nulos:', alucar.isna().sum().sum())
```

```python
alucar.dtypes
```

```python
alucar['mes'] = pd.to_datetime(alucar['mes'])
alucar.dtypes
```

```python
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
sns.lineplot(x='mes', y='vendas', data=alucar)
```

[Urban Institute Data Visualization style guide](http://urbaninstitute.github.io/graphics-styleguide/)

```python
sns.set_palette('Accent')
sns.set_style('darkgrid')
```

```python
ax = sns.lineplot(x='mes', y='vendas', data=alucar)
ax.figure.set_size_inches(12, 6)
ax.set_title('Vendas Alucar de 2017 e 2018', loc='left', fontsize=18)
ax.set_xlabel('Tempo', fontsize=14)
ax.set_ylabel('Vendas (R$)', fontsize=14)
ax = ax
```

## 02. Decomposição e autocorrelação

```python
alucar['aumento'] = alucar['vendas'].diff()
alucar.head()
```

```python
ax = sns.lineplot(x='mes', y='aumento', data=alucar)
ax.figure.set_size_inches(12, 6)
ax.set_title('Aumento das vendas Alucar de 2017 e 2018', loc='left', fontsize=18)
ax.set_xlabel('Tempo', fontsize=14)
ax.set_ylabel('Aumento', fontsize=14)
ax = ax
```

```python
def plotar(titulo, xlabel, ylabel, x, y, dataset):
    ax = sns.lineplot(x=x, y=y, data=dataset)
    ax.figure.set_size_inches(12, 6)
    ax.set_title(titulo, loc='left', fontsize=18)
    ax.set_xlabel(xlabel, fontsize=14)
    ax.set_ylabel(ylabel, fontsize=14)
    ax = ax
```
